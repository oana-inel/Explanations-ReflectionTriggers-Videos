# Generate reflection triggers on the video channel

```python
channels_df = pd.read_csv("../data/video_channels_metadata.csv")
```

```python
dataset_study["numberSubscribers"] = ""
dataset_study["creationDateChannel"] = ""
dataset_study["relatedChannels"] = ""

for i in range(len(dataset_study)):
    subset_channel = channels_df[channels_df['channel_id'] == dataset_study['channelId'].iloc[i]]
    
    dataset_study["numberSubscribers"].iloc[i] = subset_channel['channel_subscriber_count'].iloc[0]
    dataset_study["creationDateChannel"].iloc[i] = subset_channel['channel_created'].iloc[0].split("T")[0]
    #print(subset_channel['channel_featured_channels_titles'].iloc[0])
    if subset_channel['channel_featured_channels_titles'].iloc[0] != "" and str(subset_channel['channel_featured_channels_titles'].iloc[0]) != "nan":
        dataset_study["relatedChannels"].iloc[i] = subset_channel['channel_featured_channels_titles'].iloc[0].replace("[",'').replace("]",'').replace("'","")
    else:
        dataset_study["relatedChannels"].iloc[i] = "No Related Channels"

dataset_study.head(10)
```

# Overlapping Topics in Video Subtitles and Video Comments

```python
vaccination_entities = pd.read_csv("../data/video_subtitles_entities_topics/vaccination_overview_entities.csv")
freetrade_entities = pd.read_csv("../data/video_subtitles_entities_topics/freetrade_overview_entities.csv")
catalonia_entities = pd.read_csv("../data/video_subtitles_entities_topics/catalonia_overview_entities.csv")
```

```python
vaccination_entities = vaccination_entities.sort_values(['count'], ascending=False)
```

```python
freetrade_entities = freetrade_entities.sort_values(['count'], ascending=False)
```

```python
catalonia_entities = catalonia_entities.sort_values(['count'], ascending=False)
```

```python
dataset_study["allRelatedTopicsSubtitles"] = ""
entity_types = ["Location", "Person", "Organization", "Company", "HealthCondition"]

import operator


for i in range(len(dataset_study)):
    entities = pd.read_csv("../data/video_subtitles_entities/" + dataset_study["video_id"].iloc[i] + ".mp4.csv")
    topics = {}
    
    for j in range(len(entities)):
        if entities["type"].iloc[j] in entity_types:
            
            if entities["lemma"].iloc[j].replace(".", "").lower() not in [x.lower() for x in topics]:
                if dataset_study["topic"].iloc[i] == "Vaccination":
                    subset = vaccination_entities[vaccination_entities["lowercase_lemma"] == entities["lemma"].iloc[j].replace(".", "").lower()]
                    topics[entities["lemma"].iloc[j].replace(".", "")] = subset["count"].iloc[0]
                if dataset_study["topic"].iloc[i] == "Catalonia Independence":
                    subset = catalonia_entities[catalonia_entities["lowercase_lemma"] == entities["lemma"].iloc[j].replace(".", "").lower()]
                    topics[entities["lemma"].iloc[j].replace(".", "")] = subset["count"].iloc[0]
                if dataset_study["topic"].iloc[i] == "Free Trade":
                    subset = freetrade_entities[freetrade_entities["lowercase_lemma"] == entities["lemma"].iloc[j].replace(".", "").lower()]
                    topics[entities["lemma"].iloc[j].replace(".", "")] = subset["count"].iloc[0]
    
    topics = sorted(topics.items(), key=operator.itemgetter(1), reverse = True)
    
    #print(topics)
    topics_list = [x[0] for x in topics]

    dataset_study["allRelatedTopicsSubtitles"].iloc[i] = ", ".join(topics_list)
```

```python
dataset_study["allRelatedTopicsComments"] = ""
entity_types = ["Location", "Person", "Organization", "Company", "HealthCondition"]

comments_entities = pd.read_csv("../data/video_comments/comments_key_entities.csv")

for i in range(len(dataset_study)):
    
    subset = comments_entities[comments_entities["video_id"] == dataset_study["video_id"].iloc[i]]
    #print(dataset_study["video_id"].iloc[i])
    topics = list(set([str(x) for x in list(subset["lemma"])]))
    
    list_topics = []
    
    for topic in topics:
        if topic.lower() not in [x.lower() for x in list_topics]: # and topic.lower() not in list_topics:
            list_topics.append(topic)
    
    dataset_study["allRelatedTopicsComments"].iloc[i] = ", ".join(list_topics)
```

```python
dataset_study["allOverlappingTopics"] = ""
dataset_study["numberAllOverlappingTopics"] = 0

for i in range(len(dataset_study)):
    overlap = [val.lower() for val in dataset_study["allRelatedTopicsSubtitles"].iloc[i].split(", ") if val.lower() in [x.lower() for x in dataset_study["allRelatedTopicsComments"].iloc[i].split(", ")]]
    
    dataset_study["allOverlappingTopics"].iloc[i] = ", ".join(overlap)
    
    if dataset_study["allRelatedTopicsSubtitles"].iloc[i] == "" or dataset_study["allRelatedTopicsComments"].iloc[i] == "":
        dataset_study["numberAllOverlappingTopics"].iloc[i] = 0
    else:
        dataset_study["numberAllOverlappingTopics"].iloc[i] = len(overlap)
```

```python
for i in range(len(dataset_study)):
    relTopics = dataset_study["allRelatedTopicsSubtitles"].iloc[i].split(", ")
    overTopics = dataset_study["allOverlappingTopics"].iloc[i].split(", ")
    procTopics = []
    
    for relT in relTopics:
        if relT.lower() in overTopics:
            procTopics.append(relT)

    dataset_study["allOverlappingTopics"].iloc[i] = ", ".join(procTopics)
```

# Controversial Topics in Video

```python
controversial_topics = pd.read_csv("../data/controversial_wikipedia_pages.csv")

list_controversial_topics = list(controversial_topics["wikiLink"])

dataset_study["controversialTopics"] = ""
dataset_study["numberControversialTopics"] = ""

for i in range(len(dataset_study)):
    controTopics = []
    
    if dataset_study["allOverlappingTopics"].iloc[i] != "":
        topics = dataset_study["allOverlappingTopics"].iloc[i].split(", ")
        
        if dataset_study["topic"].iloc[i] == "Vaccination":            
            for topic in topics:
                subset = vaccination_entities[vaccination_entities["lowercase_lemma"] == topic.lower()]
                subset = subset["Wikipage"].iloc[0][1:-1].replace("'", "").replace("\n", "").split(" ")
                
                for t in subset:
                    if t in list_controversial_topics:
                        if topic not in controTopics:
                            controTopics.append(topic)  
        
        if dataset_study["topic"].iloc[i] == "Catalonia Independence":
            for topic in topics:
                subset = catalonia_entities[catalonia_entities["lowercase_lemma"] == topic.lower()]
                subset = subset["Wikipage"].iloc[0][1:-1].replace("'", "").replace("\n", "").split(" ")

                for t in subset:
                    if t in list_controversial_topics:
                        if topic not in controTopics:
                            controTopics.append(topic)  
                        
        if dataset_study["topic"].iloc[i] == "Free Trade":
            for topic in topics:
                subset = freetrade_entities[freetrade_entities["lowercase_lemma"] == topic.lower()]
                subset = subset["Wikipage"].iloc[0][1:-1].replace("'", "").replace("\n", "").split(" ")
                
                for t in subset:
                    if t in list_controversial_topics:
                        if topic not in controTopics:
                            controTopics.append(topic)  
        
        dataset_study["controversialTopics"].iloc[i] = ", ".join(controTopics)
        dataset_study["numberControversialTopics"].iloc[i] = len(controTopics)
    else:
        dataset_study["controversialTopics"].iloc[i] = ""
        dataset_study["numberControversialTopics"].iloc[i] = 0

        
```
