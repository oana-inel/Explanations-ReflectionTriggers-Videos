```python
processed_results_no_explanations = []

for i in range(len(results_no_explanations)):
    row = []
    row.append(results_no_explanations['HITId'].iloc[i])
    row.append(results_no_explanations['AssignmentId'].iloc[i])
    row.append(results_no_explanations['WorkerId'].iloc[i])
    row.append(results_no_explanations['Input.storageLink'].iloc[i])
    row.append(results_no_explanations['Input.video_id'].iloc[i])
    row.append(results_no_explanations['Input.topic'].iloc[i])
    row.append(results_no_explanations['Input.channelTitle'].iloc[i])
    
    if results_no_explanations['Answer.alignment.strong_agree'].iloc[i] == True:
        row.append(5)
    if results_no_explanations['Answer.alignment.agree'].iloc[i] == True:
        row.append(4)
    if results_no_explanations['Answer.alignment.neutral'].iloc[i] == True:
        row.append(3)
    if results_no_explanations['Answer.alignment.disagree'].iloc[i] == True:
        row.append(2)
    if results_no_explanations['Answer.alignment.strong_disagree'].iloc[i] == True:
        row.append(1)
        
    if results_no_explanations['Answer.alignment_topic.strong_agree'].iloc[i] == True:
        row.append(5)
    if results_no_explanations['Answer.alignment_topic.agree'].iloc[i] == True:
        row.append(4)
    if results_no_explanations['Answer.alignment_topic.neutral'].iloc[i] == True:
        row.append(3)
    if results_no_explanations['Answer.alignment_topic.disagree'].iloc[i] == True:
        row.append(2)
    if results_no_explanations['Answer.alignment_topic.strong_disagree'].iloc[i] == True:
        row.append(1)
    
    if results_no_explanations['Answer.accurate.strong_agree'].iloc[i] == True:
        row.append(5)
    if results_no_explanations['Answer.accurate.agree'].iloc[i] == True:
        row.append(4)
    if results_no_explanations['Answer.accurate.neutral'].iloc[i] == True:
        row.append(3)
    if results_no_explanations['Answer.accurate.disagree'].iloc[i] == True:
        row.append(2)
    if results_no_explanations['Answer.accurate.strong_disagree'].iloc[i] == True:
        row.append(1)  
        
    if results_no_explanations['Answer.trustworthy.strong_agree'].iloc[i] == True:
        row.append(5)
    if results_no_explanations['Answer.trustworthy.agree'].iloc[i] == True:
        row.append(4)
    if results_no_explanations['Answer.trustworthy.neutral'].iloc[i] == True:
        row.append(3)
    if results_no_explanations['Answer.trustworthy.disagree'].iloc[i] == True:
        row.append(2)
    if results_no_explanations['Answer.trustworthy.strong_disagree'].iloc[i] == True:
        row.append(1)
        
    if results_no_explanations['Answer.useful.useful_dontknow'].iloc[i] == True:
        row.append(0)
    if results_no_explanations['Answer.useful.useful_yes'].iloc[i] == True:
        row.append(1)
    if results_no_explanations['Answer.useful.useful_no'].iloc[i] == True:
        row.append(-1)   
        
    if results_no_explanations['Answer.confident.strong_agree'].iloc[i] == True:
        row.append(5)
    if results_no_explanations['Answer.confident.agree'].iloc[i] == True:
        row.append(4)
    if results_no_explanations['Answer.confident.neutral'].iloc[i] == True:
        row.append(3)
    if results_no_explanations['Answer.confident.disagree'].iloc[i] == True:
        row.append(2)
    if results_no_explanations['Answer.confident.strong_disagree'].iloc[i] == True:
        row.append(1)
        
    row.append(results_no_explanations['Answer.comments'].iloc[i])
    
    processed_results_no_explanations.append(row)
    
processed_results_no_explanations = pd.DataFrame(processed_results_no_explanations, columns=['HITId', 'AssignmentId', 'WorkerId', 'storageLink', 'video_id', 
                                          'topic', 'channelTitle', 'alignment_video', 'alignment_topic', 'accuracy_before', 
                                          'trustworthiness_before', 'usefulness_video_before', 'confidence_video_before', 'comments'])
    
processed_results_no_explanations.to_csv("processed_results_reflective_assessment_no_explanations.csv", index=False)
```

# Changes in confidence

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import mannwhitneyu
# compare samples

print("all videos")
stat, p = mannwhitneyu(processed_results_no_explanations["confidence_video_before"], processed_results["confidence_video_after"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("vaccination videos")
stat, p = mannwhitneyu(vaccination_noexpl["confidence_video_before"], vaccination["confidence_video_after"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("catalonia videos")
stat, p = mannwhitneyu(catalonia_noexpl["confidence_video_before"], catalonia["confidence_video_after"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("free trade videos")
stat, p = mannwhitneyu(freetrade_noexpl["confidence_video_before"], freetrade["confidence_video_after"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
for video_id in video_ids:
    print(video_id)
    subset1 = processed_results[processed_results["video_id"] == video_id]
    subset2 = processed_results_no_explanations[processed_results_no_explanations["video_id"] == video_id]
    stat, p = mannwhitneyu(subset2["confidence_video_before"], subset1["confidence_video_after"])
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Same distribution (fail to reject H0)')
    else:
        print('Different distribution (reject H0)')
    
```

# Changes in Explanation Usefulness

```python
from numpy.random import seed
from numpy.random import randn
from scipy.stats import mannwhitneyu
# compare samples

print("all videos")
stat, p = mannwhitneyu(resuts_within_subjects["usefulness_explanations"], processed_results["useful_explanations"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("vaccination videos")
stat, p = mannwhitneyu(vaccination_ws["usefulness_explanations"], vaccination["useful_explanations"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("catalonia videos")
stat, p = mannwhitneyu(catalonia_ws["usefulness_explanations"], catalonia["useful_explanations"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
print("free trade videos")
stat, p = mannwhitneyu(freetrade_ws["usefulness_explanations"], freetrade["useful_explanations"])
print('Statistics=%.3f, p=%.3f' % (stat, p))
# interpret
alpha = 0.05
if p > alpha:
    print('Same distribution (fail to reject H0)')
else:
    print('Different distribution (reject H0)')
    
for video_id in video_ids:
    print(video_id)
    subset1 = processed_results[processed_results["video_id"] == video_id]
    subset2 = resuts_within_subjects[resuts_within_subjects["video_id"] == video_id]
    stat, p = mannwhitneyu(subset2["usefulness_explanations"], subset1["useful_explanations"])
    print('Statistics=%.3f, p=%.3f' % (stat, p))
    # interpret
    alpha = 0.05
    if p > alpha:
        print('Same distribution (fail to reject H0)')
    else:
        print('Different distribution (reject H0)')
    
```
