```python
import sys
import six
import csv

from __future__ import print_function
from nltk.stem import *
from nltk.stem.porter import *
from nltk.stem.snowball import SnowballStemmer
from nltk.corpus import wordnet

def get_wordnet_pos(word):
    """Map POS tag to first character lemmatize() accepts"""
    tag = nltk.pos_tag([word])[0][1][0].upper()
    tag_dict = {"J": wordnet.ADJ,
                "N": wordnet.NOUN,
                "V": wordnet.VERB,
                "R": wordnet.ADV}

    return tag_dict.get(tag, wordnet.NOUN)


lemmatizer = WordNetLemmatizer()

stemmer_porter = PorterStemmer()
stemmer_snowball = SnowballStemmer('english')
stemmer_lancaster = LancasterStemmer()

videos = [PLACEHOLDER_LIST_OF_VIDEO_IDS]

names = ["storageLink","transcript","tokens","word_pos_tags","lemma","porter_stem","snowball_stem","lancaster_stem",
        "porter_stem_notprocessed","snowball_stem_notprocessed","lancaster_stem_notprocessed"]

for video in videos:

    with open("..data/video_subtitles_processed/" + video + ".csv", "w" ) as outSentences:
        writer = csv.writer( outSentences )
        writer.writerow(names)
        
    video_subtitle_file = [PLACEHOLDER_VIDEO_SUBTITLE_CSV]

    df = file.groupby(by=['storageLink','transcript','transcript_part'], as_index=False).first()
    df = df[["storageLink", "transcript", "transcript_part", "start_time", "end_time"]]
    df = df.sort_values(["transcript_part"])
    
    text = " ".join(list(df["transcript"]))
    
    if text != "":
    
        storageLink = df["storageLink"].iloc[0]
        transcript = text
        tokens = nltk.word_tokenize(text)
        
        word_pos_tags = nltk.pos_tag(tokens)
        lemma = [lemmatizer.lemmatize(w, get_wordnet_pos(w)) for w in nltk.word_tokenize(text)]
        
        porter_stem = []
        snowball_stem = []
        lancaster_stem = []
        
        for pair in word_pos_tags:
            if pair[1] == "NNP" or pair[1] == "NNPS":
                porter_stem.append(pair[0])
                snowball_stem.append(pair[0])
                lancaster_stem.append(pair[0])
            else:
                porter_stem.append(stemmer_porter.stem(pair[0]))
                snowball_stem.append(stemmer_snowball.stem(pair[0]))
                lancaster_stem.append(stemmer_lancaster.stem(pair[0]))
        
        row = [storageLink,transcript,tokens,word_pos_tags,lemma,porter_stem,snowball_stem,lancaster_stem]
        
        porter_stem = [stemmer_porter.stem(token) for token in tokens]
        snowball_stem = [stemmer_snowball.stem(token) for token in tokens]
        lancaster_stem = [stemmer_lancaster.stem(token) for token in tokens]
        
        row.append(porter_stem)
        row.append(snowball_stem)
        row.append(lancaster_stem)
        
        with open("..data/video_subtitles_processed/" + video + ".csv", "a" ) as outSentences:
            writer = csv.writer( outSentences )
            writer.writerow(row)
```
